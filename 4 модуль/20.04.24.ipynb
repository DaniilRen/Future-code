```python
import requests as re
from bs4 import BeautifulSoup
import matplotlib.pyplot as plt
```

```python
def filter_row(row):
	return ''.join([w for w in row if w.isalnum()]).title()


def process_text(raw: list) -> str:
	filtered_text = ""
	for data in raw:
		if not any([data.find('a'), data.find('b'), data.find('span')]):
			filtered_text += " " + data.text.strip()

	filtered_text = [filter_row(r) for r in filtered_text.split(" ") if r != ""]
	return filtered_text


def create_counts(text1: list, text2: list) -> list:
	word_counts = []
	for w in set(text1 + text2):
		word_counts.append((w, text1.count(w), text2.count(w)))
	
	return word_counts[1:]


url_1 = "https://guap.ru/digital"
url_2 = "https://guap.ru/digital/modul/iot"
```

```python
page = re.get(url_1)
print(page.status_code)

soup = BeautifulSoup(page.text, "html.parser")
text_raw = soup.findAll('p')
filtered_1 = process_text(text_raw)
print(filtered_1)
```

```python
page = re.get(url_2)
print(page.status_code)

soup = BeautifulSoup(page.text, "html.parser")
text_raw = soup.findAll('p')
filtered_2 = process_text(text_raw)
print(filtered_2)
```

```python
counts = create_counts(filtered_1, filtered_2)
print(counts)
```

# Задача 1
Облако слов

```python
def text_size(total: int) -> float:
	return total * 0.95

c = 1
for word, count_1, count_2 in counts:
	if c % 2 != 0:
		plt.text(count_1+15, count_2+10, word, ha="center", va="center", size=
						text_size(count_1 + count_2))
	else:
		plt.text(count_1, count_2, word, ha="center", va="center", size=
				text_size(count_1 + count_2))
	c += 1

plt.xlabel("Популярность в 1 тексте")
plt.ylabel("Популярность во 2 тексте")
plt.axis([0, 40, 0, 40])
plt.xticks([])
plt.yticks([])
plt.show()
```

```python
from typing import Dict, List
import random

Grammar = Dict[str, List[str]]

grammar = {
	"_S": ["NP VP"],
	"_NP": ["N", "A_NP_P_A_N"],
	"_VP": ["_V", "_V_NP"],
	"_N": ["learning", "C++", "Python"],
	"_A": ["linear", "regressive", "useful"],
	"_P": ["near", "about", "of"],
	"_V": ["tests", "goes", "does", "effects"]
}


def is_terminal(token: str) -> bool:
	return token[0] != "_"

def expand(grammar: Grammar, tokens: List[str]) -> List[str]:
	for i, token in enumerate(tokens):
		if is_terminal(token):
			continue
		replacement = random.choice(grammar[token])
		if is_terminal(replacement):
			tokens[i] = replacement
		else:
			tokens = tokens[:i] + replacement.split() + tokens[(i+1) :]
		return expand(grammar, tokens) 
	
	return tokens
	
def generate_sentence(grammar: Grammar) -> List[str]:
	return expand(grammar, ["_N"])


print(generate_sentence(grammar))
```
