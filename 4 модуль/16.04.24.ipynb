# Задача 2
Предсказание выживаемости человека 

```python
features = [f for f in df.columns if f != "Survived"]
x = pd.get_dummies(df[features])
y = df["Survived"]

x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.15, random_state=20)
```

```python
model = RandomForestClassifier()
model.fit(x_train, y_train)
```

```python
prediction = model.predict(x_test)
f'accuracy = {accuracy_score(y_test, prediction):.2f}, ROC = {roc_auc_score(y_test, prediction):.2f}' 
```
